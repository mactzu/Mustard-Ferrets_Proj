```python
stk_vol_d = pd.DataFrame({"Nab Close":nab_stk_vd, 
                          "CBA Close": cba_stk_vd,
                         "ANZ Close": anz_stk_vd,
                         "Westpac Close":wst_stk_vd,
                         "Maq Close": maq_stk_vd})
stk_vol_d
```

```python
market_cap = fin_day_cl * stk_vol_d 
```

```python
Total = market_cap["Nab Close"] + market_cap["CBA Close"] + market_cap["ANZ Close"]
+ market_cap["Westpac Close"] + market_cap["Maq Close"]
```

```python
weight_n = market_cap["Nab Close"] / Total
weight_c= market_cap["CBA Close"] / Total
weight_a = market_cap["ANZ Close"]/ Total
weight_w = market_cap["Westpac Close"]/ Total
weight_m = market_cap["Maq Close"]/ Total

weights = pd.DataFrame({"Nab Weight":weight_n, 
                          "CBA Weight": weight_c,
                         "ANZ Weight": weight_a,
                         "Westpac Weight":weight_w,
                         "Maq Weight": weight_m})
```

```python
daily_return = fin_day_cl.pct_change(1).dropna()
```

```python
#Weighted Ave Return
w_ave_ret = daily_return["Nab Close"]*weights["Nab Weight"] 
+ daily_return["CBA Close"]*weights["CBA Weight"]
+ daily_return["ANZ Close"]*weights["ANZ Weight"]
+ daily_return["Westpac Close"]*weights["Westpac Weight"] 
+ daily_return["Maq Close"]*weights["Maq Weight"]

w_ave_ret = w_ave_ret.rolling(window = 20).mean().dropna()
w_ave_ret
```
